```python
from bs4 import BeautifulSoup as soup
import pandas as pd
from delorean import parse
```

```python
with open('ICML_2020_data/ICML_2020_schedule.html', 'r') as inpFile:
    contents = inpFile.read()
    bsoup = soup(contents, 'lxml')
```

```python
all_schedule_list = bsoup.findAll('div', attrs={"class":"maincard"})
```

```python
schedule_events = []
for event in all_schedule_list:
    maincard_header = event.findAll('div', attrs={"class":"maincardHeader"})
    date_time_event = maincard_header[1].text
    if '&' in date_time_event:
        #print(event)
        date_times = date_time_event.split(' & ')        

        for date_time in date_times:
            event_dict = {}
            event_dict['type'] = maincard_header[0].text
            event_dict['datetime'] = date_time
            event_dict['Title'] = event.find('div', attrs={'class':'maincardBody'}).text
            event_dict['Authors/Speakers'] = ','.join(event.find('div', attrs={'class':'maincardFooter'}).text.split(' · '))
            schedule_events.append(event_dict)
            #print(schedule_events)

    else:
        event_dict = {}
        event_dict['type'] = maincard_header[0].text
        event_dict['datetime'] = date_time_event
        event_dict['Title'] = event.find('div', attrs={'class':'maincardBody'}).text
        event_dict['Authors/Speakers'] = ','.join(event.find('div', attrs={'class':'maincardFooter'}).text.split(' · '))
        schedule_events.append(event_dict)
        
    #print(event_dict)
    
```

```python
schedule_df = pd.DataFrame(schedule_events)
```

```python
schedule_df['day'] = schedule_df['datetime'].apply(lambda x:'' if (x.strip() == '') else x.split(' ')[0])
schedule_df['date'] = schedule_df['datetime'].apply(lambda x:'' if (x.strip() == '') else ' '.join(x.split(' ')[1:3]))
schedule_df['start_time'] = schedule_df['datetime'].apply(lambda x:'' if (x.strip() == '') else ' '.join(x.split(' ')[3:5]))
schedule_df['end_time'] = schedule_df['datetime'].apply(lambda x:'' if (x.strip() == '') else ' '.join(x.split(' ')[6:8]))
```
