```python
import pandas as pd
import ipywidgets as widgets
from ipywidgets import interact, interact_manual
import numpy as np
```

```python
schedule_df = pd.read_csv('ICML_2020_data/icml_schedule.csv')
```

```python
schedule_df.head()
```

```python
schedule_df.drop('datetime', inplace=True, axis=1)
```

```python
schedule_df.replace(to_replace=np.nan, value='', inplace=True)
```

```python
schedule_df.head(100)
```

```python
schedule_df.type.unique()
```

# ICML 2020 Schedule with Day filter

```python
def show_schedule_by_filters(day):
    return(schedule_df.loc[(schedule_df['day']==day)])


calendar_interact = interact(show_schedule_by_filters, 
                             day=widgets.Dropdown(options=list(schedule_df.day.unique())),
                            )
```

# ICML 2020 Schedule with Type and Date Filter

```python
def show_schedule_by_filters(type_of,date):
    return(schedule_df.loc[(schedule_df['type']==type_of) & (schedule_df['date']==date)])


calendar_interact = interact(show_schedule_by_filters, 
                             type_of=widgets.Dropdown(options=list(schedule_df.type.unique())),
                             date=widgets.Dropdown(options=list(schedule_df.date.unique())),
                            )
```
